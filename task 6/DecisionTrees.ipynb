## TASK 6  PREDICTION USING DecisionTrees

```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

## 0 means setosa , 1 means = versicolor , 2 means = virginica

### Now let us define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

## Let us visualize the Decision Tree to understand it better
## without train test split



```python
from sklearn import tree
```

```python
cn=['setosa','versicolor','virginica']
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dtree,feature_names=iris.feature_names,class_names=cn,filled=True)
plt.show()
```

## lets try with train test split 

```python
from sklearn.metrics  import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
X=df[:].values
y=iris.target
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=2)
```

```python
dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
l=iris.target_names
```

## confusion_matrix

```python
print(confusion_matrix(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,yticklabels=l,xticklabels=l)
```

## we can see setosa true 14 prediction ,versiocolor have 7 true prediction , verginica have 7 true prediction 
## setosa have 0 false prediction , versicolor have 1 false prediction , verginica 1 false prediction 

```python
confusion_matrix_check=pd.DataFrame(y_test,y_pred)
confusion_matrix_check
```

```python
px.histogram(confusion_matrix_check)
```

## in this bar we can see 14 total true prediction for setosa and versicolor have count of 8 but true is 7 and false is 1,verginica have count of 8 true is 7 and false is 1 

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
